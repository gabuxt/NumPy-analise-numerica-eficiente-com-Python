Bem-vindo(a) ao Projeto de Curso - Introdução à Biblioteca NumPy! Neste material, você encontrará um guia abrangente sobre os **principais comandos da biblioteca NumPy do Python**.

A biblioteca NumPy é uma ferramenta poderosa para a **manipulação e análise de dados numéricos** de forma eficiente. Com ela, você poderá **realizar operações matemáticas avançadas, criar e manipular arrays multidimensionais, executar cálculos estatísticos e muito mais.**

No contexto da biblioteca **NumPy**, o **import** é essencial para começarmos a utilizar suas poderosas funcionalidades. Para importar o NumPy em nosso código, utilizamos a seguinte linha:

```python
import numpy as np
```


Essa linha de código define uma variável chamada url que armazena uma string contendo o endereço de um **arquivo CSV**.

```python
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv'
```

Essa linha de código utiliza a função **loadtxt()** da biblioteca NumPy para carregar os dados de um arquivo CSV localizado na URL especificada pela variável url. O parâmetro **delimiter=','** indica que o arquivo CSV está separado por vírgulas.

Em seguida, o parâmetro **usecols=np.arange(1, 88, 1)** é utilizado para selecionar quais colunas do arquivo CSV serão carregadas. Neste caso, a função np.arange(1, 88, 1) retorna um array NumPy contendo os números de 1 a 87, especificando assim que queremos carregar as colunas de 1 a 87 do arquivo CSV.

```python
dado = np.loadtxt(url,delimiter=',', usecols=np.arange(1,88,1))
```

Essa linha de código retorna uma tupla que descreve a **forma ou o tamanho do array** dado. O atributo **shape** é utilizado em arrays NumPy para obter informações sobre a estrutura do array, ou seja, a **quantidade de elementos em cada dimensão**.

```python
dado.shape
```

Essa linha de código retorna a** matriz transposta** do array dado. A transposição de uma matriz envolve a **troca das linhas pelas colunas e vice-versa**.

Ao executar **dado.T**, é retornado um novo array que contém os mesmos valores de dado, porém com as linhas e colunas trocadas. Isso significa que os elementos que estavam originalmente nas linhas do dado agora estão nas colunas do array resultante e vice-versa.

```python
dado_transposto = dado.T
```

Essa linha de código retorna a **primeira coluna** do array dado_transposto[:,O] **:** antes da vírgula indica que queremos selecionar todas as linhas do array, enquanto o **0** após a vírgula indica que queremos selecionar a **coluna de índice 0**.

```python
datas = dado_transposto[:,0]
```

Essa linha de código retorna a primeira coluna do array dado_transposto[:,O] : antes da vírgula indica que queremos selecionar todas as linhas do array, enquanto o 1:6 após a vírgula indica que queremos selecionar as coluna de índice 1 até o índice 5, pois o ultimo indice é excluido.

Essas linhas de código são úteis para visualizar e comparar os dados de cada ano separadamente, permitindo identificar padrões ou tendências ao longo dos meses em cada ano específico. A legenda adicionada facilita a interpretação dos diferentes conjuntos de dados no gráfico.

Essa linha de código verifica se os arrays moscow_ano3 e moscow_ano4 são iguais, ou seja, se possuem exatamente os **mesmos elementos nos mesmos índices**.

Ao executar np.array_equal(moscow_ano3, moscow_ano4), a função **array_equal()** do NumPy é utilizada para **comparar** os dois arrays. Se os arrays forem iguais, a função retornará **True**, caso contrário, retornará **False**.

```python
np.array_equal(moscow_ano3, moscow_ano4)
```

Essa linha de código verifica se os arrays moscow_ano3 e moscow_ano4 são **aproximadamente iguais** dentro de uma **tolerância específica**.

Ao executar np.allclose(moscow_ano3, moscow_ano4, 10), a função **allclose()** do NumPy é utilizada para comparar os **dois arrays**. O terceiro argumento, 10, **representa a tolerância**, ou seja, o **valor máximo permitido para a diferença entre os elementos** dos arrays.

Se a diferença absoluta entre os elementos dos arrays moscow_ano3 e moscow_ano4 for menor ou igual à tolerância especificada (10 neste caso), a função retornará **True**. Caso contrário, retornará **False**.

```python
np.allclose(moscow_ano3, moscow_ano4, 10)
```

image.png

```python
Y = moscow
X = datas
n = np.size(moscow)
```

```python
a = (n*np.sum(X*Y) - np.sum(X) * np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
```

Essa linha de código gera um array de **100 números inteiros aleatórios**, dentro de um **intervalo especificado.**

Ao executar np.random.randint(low=40, high=100, size=100), a função **np.random.randint** do NumPy é utilizada para gerar **números inteiros aleatórios**. Os argumentos passados são:

**low**=40: Define o valor mínimo (inclusive) do intervalo dos números gerados, neste caso, 40.
**high**=100: Define o valor máximo (exclusivo) do intervalo dos números gerados, neste caso, 100.
**size**=100: Define o tamanho do array gerado, ou seja, a quantidade de números inteiros aleatórios desejada, neste caso, 100.
Dessa forma, a linha de código cria um array com 100 números inteiros aleatórios, cada um dentro do intervalo de 40 a 99.

```python
dados = np.random.randint(low=40, high=100, size = 100)
```

Essa linha de código salva um array em **formato CSV** em um arquivo chamado 'dados.csv'.

Ao executar np.savetxt('dados.csv', dados, delimiter=','), a função **np.savetxt** do NumPy é utilizada para salvar o array no formato CSV. Os argumentos passados são:

**'dados.csv'**: O nome do arquivo de saída onde o array será salvo.
**dados**: O array que será salvo no arquivo.
**delimiter=','**: O caractere delimitador que separa os valores no arquivo CSV, neste caso, a vírgula ','.

Dessa forma, a linha de código cria um arquivo chamado 'dados.csv' e salva o conteúdo do array no formato CSV, com os valores separados por vírgulas.

```python
np.savetxt('dados.csv', dados, delimiter=',')
```
